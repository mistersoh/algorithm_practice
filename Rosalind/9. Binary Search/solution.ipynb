```python

with open('test.txt','r') as f:
    sample = f.read().splitlines()

list_A = [int(i) for i in sample[2].split()]
list_B = [int(i) for i in sample[3].split()]

def Binary_search(Array, item):
    low = 0
    high = len(Array)-1
    while low <= high:
        index = (low + high) // 2
        if item == Array[index]:
            return index + 1
        elif item > Array[index]:
            low = index + 1
        else:
            high = index - 1
    return -1

res = ''
for i in list_B:
    res += str(Binary_search(list_A,i)) + ' '

res
```
