```python
from collections import Counter, defaultdict

with open('test.txt','r') as f:
    s = f.read().splitlines()


dict_s = defaultdict(int)
tuple_s = []
for i in s:
    tup = eval(i.replace(' ', ','))
    tuple_s.append(tup)
    dict_s[tup[0]] += 1
    dict_s[tup[1]] += 1

solos = []
for i,j in dict_s.items():
    if j == 1:
        solos.append(i)

for i in solos:
    for tup in tuple_s:
        if i in tup:
            print(tup)
            dict_s[tup[0]] -= 1
            dict_s[tup[1]] -= 1

res = ''
    
for i in sorted(dict_s):
    if dict_s[i] != 0:
        res += str(dict_s[i]) + ' '

res
```
